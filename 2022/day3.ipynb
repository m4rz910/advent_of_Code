```python
import pandas as pd
import numpy as np
```

```python
letters = [chr(i) for i in range(97,123)] + [chr(i) for i in range(65,91)]
letter_vals = {letters[i-1]:i for i in range(1,len(letters)+1)}
```

# part 1

```python
def find_common_item(comp1,comp2):
    common_items = []
    for i, elements in enumerate(comp1):
        for j,e in enumerate(elements):
            if e in comp2[i]:
                common_items.append(e)
                break
                                    
    return common_items

df = pd.read_csv('data/day3.csv', header=None)
df['comp1'] = df.apply(lambda s: s[0][:len(s[0])//2], axis='columns')
df['comp2'] = df.apply(lambda s: s[0][len(s[0])//2:], axis='columns')

df['common_items'] = find_common_item(comp1=df['comp1'].values,
                                      comp2=df['comp2'].values)
df['value'] = df.apply(lambda s: letter_vals[s['common_items']], axis='columns')
df
```
