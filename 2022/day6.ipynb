```python
import pandas as pd
import numpy as np
```

```python
buff = pd.read_csv('data/day6.csv',header=None).iloc[0,0]
#buff = "zcfzfwzzqfrljwzlrfnpqdbhtmscgvjw"
buff = pd.Series(data=[c for c in buff], index=range(1,len(buff)+1))
buff
```

# part 1

```python
series = [x for x in buff.rolling(4) if len(x)==4]

for s in series:
    if len(s.values) == len(set(s.values)):
        print(s)
```
