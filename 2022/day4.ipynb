```python
import pandas as pd
```

```python
df = pd.read_csv("data/day4.csv", header=None)

# key lesson: make sure you convert to the data type you need to use int in this case
df['a1_l'] = df.apply(lambda s: int(s[0].split(",")[0].split("-")[0]), axis='columns') #assigment1 lower bound
df['a1_u'] = df.apply(lambda s: int(s[0].split(",")[0].split("-")[1]), axis='columns') #assigment1 upper
df['a2_l'] = df.apply(lambda s: int(s[0].split(",")[1].split("-")[0]), axis='columns') #assigment2 lower bound
df['a2_u'] = df.apply(lambda s: int(s[0].split(",")[1].split("-")[1]), axis='columns') #assigment2 upper
df
```

# part 1

```python
def f(s):
    if (s['a1_l']>=s['a2_l']) and (s['a1_u']<=s['a2_u']):
        return 'a1_in_a2'
    elif (s['a1_l']<=s['a2_l']) and (s['a1_u']>=s['a2_u']):
        return 'a2_in_a1'
    else:
        return 'None'

df['completeoverlap'] = df.apply(lambda s: f(s), axis='columns') 
df
```

```python
count = df['completeoverlap'].value_counts()
count[[v for v in count.index if v != 'None']].sum()
```

# part 2

```python
def f(s):
    if (s['a1_l']<=s['a2_l']) and (s['a1_u']>=s['a2_l']):
        return 'at_least_some_overlap'
    elif (s['a2_l']<=s['a1_l']) and (s['a2_u']>=s['a1_l']):
        return 'at_least_some_overlap'
    else:
        return 'None'
    
df['at_least_some_overlap'] = df.apply(lambda s: f(s), axis='columns') 
df
```

```python
df['at_least_some_overlap'].value_counts()
```
